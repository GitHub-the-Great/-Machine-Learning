```python
import numpy as np
import matplotlib.pyplot as plt
import struct

# ========== Load Dataset ==========
def load_images(filename):
    with open(filename, 'rb') as f:
        _, num, rows, cols = struct.unpack(">IIII", f.read(16))
        data = np.frombuffer(f.read(), dtype=np.uint8).reshape((num, rows * cols))
    return data.astype(np.float32) / 255.0

def load_labels(filename):
    with open(filename, 'rb') as f:
        _, num = struct.unpack(">II", f.read(8))
        labels = np.frombuffer(f.read(), dtype=np.uint8)
    return labels

# ========= 2. Utils ==========
def shuffle_numpy(X, y):
    """Shuffle X and y in the same order"""
    indices = np.random.permutation(len(X))
    return X[indices], y[indices]

def split_train_val(X, y, val_ratio=0.2):
    """Split the dataset into training and validation sets"""
    # Convert binary classification problem (digits 0 and 1 only)
    mask = (y == 0) | (y == 1)
    X_binary = X[mask]
    y_binary = y[mask]
    
    # Shuffle and split
    X_shuffled, y_shuffled = shuffle_numpy(X_binary, y_binary)
    val_size = int(len(X_shuffled) * val_ratio)
    
    X_train = X_shuffled[val_size:]
    y_train = y_shuffled[val_size:]
    X_val = X_shuffled[:val_size]
    y_val = y_shuffled[:val_size]
    
    return X_train, y_train, X_val, y_val

def one_hot(y, num_classes):
    """Convert labels to one-hot encoding"""
    one_hot_y = np.zeros((y.size, num_classes))
    one_hot_y[np.arange(y.size), y] = 1
    return one_hot_y

def accuracy(Y_pred, Y_true):
    """Calculate accuracy from predictions and true labels"""
    if len(Y_true.shape) > 1:  # One-hot encoded
        Y_true_labels = np.argmax(Y_true, axis=1)
    else:
        Y_true_labels = Y_true
    
    Y_pred_labels = np.argmax(Y_pred, axis=1)
    return np.mean(Y_pred_labels == Y_true_labels)

# ========= 3. Model ==========
class MLP:
    def __init__(self, input_dim, hidden_dim, output_dim, weight_decay=0.0):
        self.W1 = np.random.randn(input_dim, hidden_dim) * 0.01
        self.b1 = np.zeros((1, hidden_dim))
        self.W2 = np.random.randn(hidden_dim, output_dim) * 0.01
        self.b2 = np.zeros((1, output_dim))
        self.lambda_ = weight_decay
        
    def relu(self, x):
        return np.maximum(0, x)
    
    def relu_deriv(self, x):
        return (x > 0).astype(float)
    
    def softmax(self, x):
        exps = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exps / np.sum(exps, axis=1, keepdims=True)
    
    def forward(self, X):
        self.z1 = X @ self.W1 + self.b1
        self.a1 = self.relu(self.z1)
        self.z2 = self.a1 @ self.W2 + self.b2
        self.a2 = self.softmax(self.z2)
        return self.a2
    
    def compute_loss(self, Y_pred, Y_true):
        """Cross-entropy loss with L2 regularization"""
        m = Y_true.shape[0]
        # Cross-entropy loss
        log_likelihood = -np.sum(Y_true * np.log(Y_pred + 1e-9)) / m
        
        # L2 regularization term
        l2_term = 0.5 * self.lambda_ * (np.sum(np.square(self.W1)) + np.sum(np.square(self.W2)))
        
        # Total loss
        return log_likelihood + l2_term
    
    def backward(self, X, Y_true, Y_pred, lr=0.1):
        m = Y_true.shape[0]
        dz2 = (Y_pred - Y_true) / m
        dW2 = self.a1.T @ dz2 + self.lambda_ * self.W2
        db2 = np.sum(dz2, axis=0, keepdims=True)
        
        da1 = dz2 @ self.W2.T
        dz1 = da1 * self.relu_deriv(self.z1)
        dW1 = X.T @ dz1 + self.lambda_ * self.W1
        db1 = np.sum(dz1, axis=0, keepdims=True)
        
        self.W2 -= lr * dW2
        self.b2 -= lr * db2
        self.W1 -= lr * dW1
        self.b1 -= lr * db1

# ========= 4. Train Function ==========
def train(model, X_train, y_train, X_val, y_val, lr=0.1, epochs=100, use_early_stopping=False, patience=5):
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    best_val_loss = np.inf
    patience_count = 0
    
    for epoch in range(epochs):
        # Forward pass on training data
        y_train_pred = model.forward(X_train)
        train_loss = model.compute_loss(y_train_pred, y_train)
        train_acc = accuracy(y_train_pred, y_train)
        
        # Backward pass (update weights)
        model.backward(X_train, y_train, y_train_pred, lr=lr)
        
        # Forward pass on validation data (no weight updates)
        y_val_pred = model.forward(X_val)
        val_loss = model.compute_loss(y_val_pred, y_val)
        val_acc = accuracy(y_val_pred, y_val)
        
        # Store metrics
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_accs.append(train_acc)
        val_accs.append(val_acc)
        
        print(f"Epoch {epoch:02d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
        
        # Early stopping check
        if use_early_stopping:
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                patience_count = 0
            else:
                patience_count += 1
            
            if patience_count >= patience:
                print(f"Early stopping triggered after {epoch+1} epochs")
                break
    
    return train_losses, val_losses, train_accs, val_accs

# ========= 5. Plotting ==========
def plot_curves(train_losses, val_losses, train_accs, val_accs, title):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Val Loss")
    plt.title("Loss Curve - " + title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label="Train Acc")
    plt.plot(val_accs, label="Val Acc")
    plt.title("Accuracy Curve - " + title)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    
    plt.tight_layout()
    plt.savefig(f"{title.replace(' ', '_')}.png")
    plt.show()

# ========= 6. Main ==========
if __name__ == "__main__":
    # Load data
    X = load_images("train-images.idx3-ubyte_____")
    y = load_labels("train-labels.idx1-ubyte_____")
    
    # Shuffle data
    X, y = shuffle_numpy(X, y)
    
    # Split data into train/val sets
    X_train, y_train, X_val, y_val = split_train_val(X, y)
    
    # One-hot encode the labels
    y_train_oh = one_hot(y_train, 2)  # Binary classification (0 and 1)
    y_val_oh = one_hot(y_val, 2)      # Binary classification (0 and 1)
    
    # Model parameters
    input_dim = 784  # 28x28 pixels
    hidden_dim = 128  # Number of hidden units
    output_dim = 2   # Binary classification
    
    # === OPTION 1: Baseline (No Regularization) ===
    print("Training baseline model (no regularization)...")
    model_base = MLP(input_dim, hidden_dim, output_dim, weight_decay=0.0)
    t_base, v_base, a_base, va_base = train(
        model_base, X_train, y_train_oh, X_val, y_val_oh, 
        lr=0.1, epochs=100, use_early_stopping=False)
    plot_curves(t_base, v_base, a_base, va_base, title="Baseline (No Regularization)")
    
    # === OPTION 2: Early Stopping ===
    print("\nTraining model with early stopping...")
    model_early = MLP(input_dim, hidden_dim, output_dim, weight_decay=0.0)
    t_early, v_early, a_early, va_early = train(
        model_early, X_train, y_train_oh, X_val, y_val_oh, 
        lr=0.1, epochs=100, use_early_stopping=True, patience=5)
    plot_curves(t_early, v_early, a_early, va_early, title="Early Stopping")
    
    # === OPTION 3: L2 Weight Decay (try different lambda values) ===
    lambda_values = [0.001, 0.0055, 0.01, 0.055, 0.1]
    
    for lambda_val in lambda_values:
        print(f"\nTraining model with weight decay (lambda={lambda_val})...")
        model_l2 = MLP(input_dim, hidden_dim, output_dim, weight_decay=lambda_val)
        t_l2, v_l2, a_l2, va_l2 = train(
            model_l2, X_train, y_train_oh, X_val, y_val_oh, 
            lr=0.1, epochs=100, use_early_stopping=False)
        plot_curves(t_l2, v_l2, a_l2, va_l2, title=f"Weight Decay (lambda={lambda_val})")
    
    # === OPTION 4: Combining Early Stopping and Weight Decay ===
    print("\nTraining model with early stopping and weight decay...")
    model_combined = MLP(input_dim, hidden_dim, output_dim, weight_decay=0.01)
    t_combined, v_combined, a_combined, va_combined = train(
        model_combined, X_train, y_train_oh, X_val, y_val_oh, 
        lr=0.1, epochs=100, use_early_stopping=True, patience=5)
    plot_curves(t_combined, v_combined, a_combined, va_combined, title="Early Stopping + Weight Decay")
```
